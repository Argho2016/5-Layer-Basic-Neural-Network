```python
from PIL import Image
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
path1= r'E:/aiub/MSCS/Datasets/Final/6000200094514.jpg'
path2= r'E:/aiub/MSCS/Datasets/Final/resized/6000200094514.jpg'
```

```python
apple = np.array(Image.open(path1).resize((64,64)))
Image.fromarray(apple).save(path2)
print("After resizing:", apple.shape)
plt.imshow(apple)
```

```python
data = np.asarray(apple)       
print(type(data))
print(data.shape)
```

```python
apple_norm = image = data.astype(float) / 255.0
#print(apple_norm)
```

```python
def convolution2d(image, kernel):
    m, n = kernel.shape
    if (m == n):
        y, x, z = image.shape
        y = y - m + 1
        x = x - n + 1
        new_image = np.zeros((y,x,z))
        for i in range(y):
            for j in range(x):
                for k in range(z):
                    new_image[i][j][k] = np.sum(image[i:i+m, j:j+n, k] * kernel)

    return new_image
```

```python
my_filter1 = np.array([
    [0, -1, 0],                    
    [-1, 5, -1],
    [0, -1, 0]
])
#SHARPEN
conv_image1 = convolution2d(apple_norm, my_filter1)
#plt.imshow(conv_image1)
print(conv_image1.shape)

#ReLU
conv_max1 = np.maximum(conv_image1, 0)
plt.figure()

f, axarr = plt.subplots(3,1) 

axarr[0].imshow(apple)
axarr[1].imshow(conv_image1)
axarr[2].imshow(conv_max1)

axarr[1].set_title('After Convolution')
axarr[0].set_title('Original')
axarr[2].set_title('After ReLU')
plt.tight_layout()

```
